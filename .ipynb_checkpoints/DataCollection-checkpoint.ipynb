```python
genres=set()
for item in mov_large_copy.genres:
    if(type(item)==str):
        x=re.finditer("(\'name\':\s\')(?P<name>\w*)(?=\')",item)
        for s in x:
            genres.add(s.groupdict()['name'])
print(genres)
```

```python
set_actors=set()
for item in mov_large_copy.cast:
    if(type(item)==str):
        x=re.finditer("(\'name\':\s\')(?P<name>[\w\.\s\-]*)(?=\')",item)
        for s in x:
            set_actors.add(s.groupdict()['name'])
print(set_actors)
```

```python
len(set_actors)
```

```python
actors=pd.DataFrame(columns=['MoviesN','AvgPop','AvgIMDB'],index=set_actors)
actors['MoviesN']=0
for j in mov_large_copy.index:
    cast=mov_large_copy.cast[j]
    if(type(cast)==str):
        x=re.findall("(\'name\':\s\')(?P<name>[\w\s\.\-]*)(?=\')",cast)
        for elem in x:
            count=actors.loc[elem[1],'MoviesN']
            if not count==0:
                if not math.isnan(mov_large_copy.popularity[j]) and not math.isnan(mov_large_copy.imdb_rating[j]):
                    actors.loc[elem[1],'MoviesN']=count+1
                    actors.loc[elem[1],'AvgPop']=(actors.loc[elem[1],'AvgPop']*count+mov_large_copy.popularity[j])/(count+1)
                    actors.loc[elem[1],'AvgIMDB']=(actors.loc[elem[1],'AvgIMDB']*count+mov_large_copy.imdb_rating[j])/(count+1)
            else:
                if not math.isnan(mov_large_copy.popularity[j]) and not math.isnan(mov_large_copy.imdb_rating[j]):
                    actors.loc[elem[1],'MoviesN']=1
                    actors.loc[elem[1],'AvgPop']=mov_large_copy.popularity[j]
                    actors.loc[elem[1],'AvgIMDB']=mov_large_copy.imdb_rating[j]
actors=actors.rename_axis('name').reset_index()
```

```python
actors.head()
```

```python
actors['Score']=0
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler2=MinMaxScaler()
scaler2.fit(actors[['AvgPop','AvgIMDB']])
actors[['AvgPop','AvgIMDB']]=scaler2.transform(actors[['AvgPop','AvgIMDB']])
actors=actors.sort_values(by=['MoviesN','AvgPop','AvgIMDB'],ascending=False)
```

```python
for j in actors.index:
    actors.loc[j,'Score']=1/(1+math.exp(-actors.loc[j,'MoviesN']+10))*(0.3*actors.loc[j,'AvgPop']+0.7*actors.loc[j,'AvgIMDB'])
```

```python
actors=actors.sort_values(by=['Score'],ascending=False)
```

```python
actors.head()
```

```python
with open("./DataSets/actors.csv",'w') as file:
    actors.to_csv(file,index=False)
```
